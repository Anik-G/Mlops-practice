```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Hey {self.owner}, {amount} is deposited. New balance is {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Hey {self.owner}, Insufficient funds!")
        else:
            self.balance -= amount
            print(f"Hey {self.owner}, {amount} is withdrawn. New balance is {self.balance}")

    def get_balance(self):
        return self.balance


# 🔹 Store multiple accounts
accounts = {
    "Anik": BankAccount("Anik", 5000),
    "Anik_2": BankAccount("Anik_2", 3000),
    "Anik_3": BankAccount("Anik_3", 7000)
}


# 🔹 Function to check balance
def check_balance(owner_name):
    try:
        balance = accounts[owner_name].get_balance()
        print(f"{owner_name}, your balance is {balance}")
    except KeyError:
        print(f"User {owner_name} does not exist")
    finally:
        print("Transaction Operation completed")


# 🔹 Take input from user
name = input("Enter account holder name: ")
check_balance(name)

# If you want “default try-catch-finally everywhere”:
def safe_operation(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print("Error:", e, "Invalid operation")
        finally:
            print("banking Operation completed")
    return wrapper

```

```python
import os
import csv

# 🔹 Decorator (global error handler)
def safe_operation(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except KeyError:
            print("User does not exist")
        except Exception as e:
            print("Error:", e, "Invalid operation")
        finally:
            print("Banking Operation completed")
    return wrapper


class BankAccount:
    def __init__(self, owner, balance=0, pin=None):
        self.owner = owner
        self.balance = balance
        self.pin = pin

    def open_account():
        owner = input("Enter new account holder name: ")
        if owner in accounts:
            print("Account already exists")
            return
        pin = int(input("Set PIN: "))
        if not (pin.is_integer() and len(str(pin)) == 4):
            print("Invalid PIN format. PIN should be a 4-digit number.")
            return
        balance = float(input("Enter initial balance: "))
        accounts[owner] = BankAccount(owner, balance, pin)
        print(f"Account created successfully for {owner}, Balance: {balance}, PIN set.")
        save_accounts_to_csv()

    def _validate_pin(self, pin):
        if self.pin != pin:
            print("Invalid PIN")
            return False
        elif not (pin.is_integer() and len(str(pin)) == 4):
            print("Invalid PIN format. PIN should be a 4-digit number.")
            return False
        return True
    
    @safe_operation
    def change_pin(self, pin, new_pin):
        self._validate_pin(pin)   # will raise error if wrong
        self.pin = new_pin
        save_accounts_to_csv()
        print("PIN changed successfully")
            
    def deposit(self, amount, pin):
        if not self._validate_pin(pin):
            return
        self.balance += amount
        print(f"Hey {self.owner}, {amount} is deposited. New balance is {self.balance}")
        save_accounts_to_csv()

    def withdraw(self, amount, pin):
        if not self._validate_pin(pin):
            return
        if amount > self.balance:
            print(f"Hey {self.owner}, Insufficient funds!")
        else:
            self.balance -= amount
            print(f"Hey {self.owner}, {amount} is withdrawn. New balance is {self.balance}")
            save_accounts_to_csv()

    def get_balance(self, pin):
        if not self._validate_pin(pin):
            return
        print(f"{self.owner}, your balance is {self.balance}")

# 🔹 Load accounts from CSV
accounts = {}
file_path = 'accounts.csv'
try:
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            owner = row['owner']
            balance = float(row['balance'])
            pin = row['pin']
            accounts[owner] = BankAccount(owner, balance, pin)
except FileNotFoundError:
    print(f"accounts.csv file not found in the directory {os.getcwd()}")
# 🔹 Save accounts back to CSV
def save_accounts_to_csv():
    with open(file_path, 'w', newline='') as file:
        fieldnames = ['owner', 'balance', 'pin']
        writer = csv.DictWriter(file, fieldnames=fieldnames)

        writer.writeheader()
        for acc in accounts.values():
            writer.writerow({
                'owner': acc.owner,
                'balance': acc.balance,
                'pin': acc.pin
            })

# 🔹 Function to perform actions
@safe_operation
def perform_action():
    action = input("Choose action (open/deposit/withdraw/check/change_pin): ").lower()
    if action == "open":
        BankAccount.open_account()
        return
    
    name = input("Enter account holder name: ")
    pin = int(input("Enter PIN: "))
    if not (pin.is_integer() and len(str(pin)) == 4):
        print("Invalid PIN format. PIN should be a 4-digit number.")
        return

    if action in ["deposit", "withdraw"]:
        amount = float(input("Enter amount: "))
        
    account = accounts[name]  # KeyError handled by decorator

    if action == "deposit":
        account.deposit(amount, pin)

    elif action == "withdraw":
        account.withdraw(amount, pin)

    elif action == "check":
        account.get_balance(pin)
    
    elif action == "change_pin":
        new_pin = int(input("Enter new PIN: "))
        account.change_pin(pin, new_pin)

    else:
        print("Invalid action")
# 🔹 Run
perform_action()
```

```python
import os
import csv
import random

# 🔹 Decorator (global error handler) ref 9.3-Decorators.ipynb
def safe_operation(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except KeyError:
            print("User does not exist")
        except Exception as e:
            print("Error:", e, "Invalid operation")
        finally:
            print("Banking Operation completed")
    return wrapper


class BankAccount:
    def __init__(self, owner, balance=0, pin=None, customer_id=None, retry=0, is_locked=False):
        self.owner = owner
        self.balance = balance
        self.pin = pin
        self.customer_id = customer_id
        self.retry = retry
        self.is_locked = is_locked

    def open_account():
        owner = input("Enter new account holder name: ")
        if owner in accounts:
            print("Account already exists")
            return
        pin = int(input("Set PIN: "))
        if not (pin.is_integer() and len(str(pin)) == 4):
            print("Invalid PIN format. PIN should be a 4-digit number.")
            return
        balance = float(input("Enter initial balance: "))
        customer_id = random.randint(10000, 999999)
        
        accounts[owner] = BankAccount(owner, balance, pin, customer_id, retry=0, is_locked=False)
        print(f"Account created successfully for {owner}, Balance: {balance}, PIN set. Customer ID: {customer_id}")
        save_accounts_to_csv()

    def _validate_pin(self, pin):
        if self.is_locked:
            print("Account is locked due to multiple failed attempts. Please contact support.")
            return False
        if self.pin != str(pin):
            self.retry+=1
            print(f"Invalid PIN. Attempts remaining: {3 - self.retry}")
            if self.retry >= 3:
                self.is_locked = True
            save_accounts_to_csv()
            return False
        elif not (pin.is_integer() and len(str(pin)) == 4):
            print("Invalid PIN format. PIN should be a 4-digit number.")
            return False
        
        self.retry = 0   # If PIN is correct, reset retry count
        save_accounts_to_csv()
        return True
    
    @safe_operation
    def change_pin(self, pin, new_pin):
        self._validate_pin(pin)   # will raise error if wrong
        self.pin = new_pin
        save_accounts_to_csv()
        print("PIN changed successfully")
            
    def deposit(self, amount, pin):
        if not self._validate_pin(pin):
            return
        self.balance += amount
        print(f"Hey {self.owner}, {amount} is deposited. New balance is {self.balance}")
        save_accounts_to_csv()

    def withdraw(self, amount, pin):
        if not self._validate_pin(pin):
            return
        if amount > self.balance:
            print(f"Hey {self.owner}, Insufficient funds!")
        else:
            self.balance -= amount
            print(f"Hey {self.owner}, {amount} is withdrawn. New balance is {self.balance}")
            save_accounts_to_csv()

    def get_balance(self, pin):
        if not self._validate_pin(pin):
            return
        print(f"{self.owner}, your balance is {self.balance}")

# 🔹 Load accounts from CSV
accounts = {}
file_path = 'accounts_1.csv'
try:
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            owner = row['owner']
            balance = float(row['balance'])
            pin = row['pin']
            customer_id = row['customer_id']
            retry = int(row['retry'])
            is_locked = row['is_locked'] == 'True'
            accounts[owner] = BankAccount(owner, balance, pin, customer_id, retry, is_locked)
except FileNotFoundError:
    print(f"{file_path} file not found in the directory {os.getcwd()}")
    save_accounts_to_csv()
    print(f"{file_path} file created in the directory {os.getcwd()}")
# 🔹 Save accounts back to CSV
def save_accounts_to_csv():
    with open(file_path, 'w', newline='') as file:
        fieldnames = ['owner', 'balance', 'pin', 'customer_id', 'retry', 'is_locked']
        writer = csv.DictWriter(file, fieldnames=fieldnames)

        writer.writeheader()
        for acc in accounts.values():
            writer.writerow({
                'owner': acc.owner,
                'balance': acc.balance,
                'pin': acc.pin,
                'customer_id': acc.customer_id,
                'retry': acc.retry,
                'is_locked': acc.is_locked
            })

# 🔹 Function to perform actions
@safe_operation
def perform_action():
    action = input("Choose action (open/deposit/withdraw/check/change_pin): ").lower()
    if action == "open":
        BankAccount.open_account()
        return
    
    name = input("Enter account holder name: ")
    pin = int(input("Enter PIN: "))
    if not (pin.is_integer() and len(str(pin)) == 4):
        print("Invalid PIN format. PIN should be a 4-digit number.")
        return

    if action in ["deposit", "withdraw"]:
        amount = float(input("Enter amount: "))
        
    account = accounts[name]  # KeyError handled by decorator

    if action == "deposit":
        account.deposit(amount, pin)

    elif action == "withdraw":
        account.withdraw(amount, pin)

    elif action == "check":
        account.get_balance(pin)
    
    elif action == "change_pin":
        new_pin = int(input("Enter new PIN: "))
        account.change_pin(pin, new_pin)

    else:
        print("Invalid action")
# 🔹 Run
perform_action()
```
