#### Abstraction
Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort.

```python
from abc import ABC,abstractmethod

## Abstract base cclass
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car enginer started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car=Car()
operate_vehicle(car)

```

```python
from abc import ABC,abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

    def sleep(self):
        animal = self.__class__.__name__
        print(f"{animal} is sleeping")

class Dog(Animal):
    def make_sound(self):
        print("Bark")

class Cat(Animal):
    def make_sound(self):
        print("Meow")

def interact_with_animal(animal):
    animal.make_sound()
    animal.sleep()

dog = Dog()
cat = Cat()

interact_with_animal(dog)
interact_with_animal(cat)
```

```python
from abc import ABC,abstractmethod

class PaymentProcessor(ABC):
    def authorize(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be positive")
        return self._authorize(amount)

    @abstractmethod
    def _authorize(self, amount):
        pass

    @abstractmethod
    def capture(self, transaction_id):
        pass

    def refund(self, transaction_id):
        print(f"Refunding transaction {transaction_id}")
        return True


class StripeProcessor(PaymentProcessor):
    def _authorize(self, amount):
        print(f"Authorizing ${amount:.2f} with Stripe")
        return f"stripe_txn_{int(amount * 1000)}"

    def capture(self, transaction_id):
        print(f"Capturing Stripe transaction {transaction_id}")
        return True


class PayPalProcessor(PaymentProcessor):
    def _authorize(self, amount):
        print(f"Authorizing ${amount:.2f} with PayPal")
        return f"paypal_txn_{int(amount * 1000)}"

    def capture(self, transaction_id):
        print(f"Capturing PayPal transaction {transaction_id}")
        return True


def process_order(order, processor: PaymentProcessor):
    transaction_id = processor.authorize(order["amount"])
    processor.capture(transaction_id)
    return transaction_id


order = {"id": 1001, "amount": 49.99}
stripe_processor = StripeProcessor()
paypal_processor = PayPalProcessor()

print(process_order(order, stripe_processor))
print(process_order(order, paypal_processor))
```
