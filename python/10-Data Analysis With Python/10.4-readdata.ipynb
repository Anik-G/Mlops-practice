```python
from urllib.request import Request, urlopen
from io import BytesIO

stata_url = 'https://www.stata-press.com/data/r16/breathe.dta'
req = Request(stata_url, headers={'User-Agent': 'curl/7.64.1'})
with urlopen(req) as resp:
    df_stata = pd.read_stata(BytesIO(resp.read()))

df_stata.head()
```
